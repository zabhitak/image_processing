```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plot

import cv2
```

```python
frame = cv2.imread("iron_man.jpg")
```

```python
plot.imshow(frame) #change rgb to bgr
```

```python
rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

```python
plot.imshow(rgb)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
face_detection = classifier.detectMultiScale(frame)
```

```python
face = face_detection[0]
```

```python
x,y,w,h = face
```

```python
face = rgb[y:y+h,x:x+h]
```

```python
plot.imshow(face)
```

```python
out  = cv2.rectangle(rgb,(x,y),(x+w,y+h),(200,0,0),4)
```

```python
plot.imshow(out)
```
